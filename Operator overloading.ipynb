### Bài 1:
Một đơn thức theo biến x được xác định bởi hệ số và số mũ, P(x) = a*x^b
Hãy xây dựng lớp biểu diễn khái niệm đơn thức cho phép thực hiện các yêu cầu sau:
- Khởi tạo đơn thức khi biết hệ số và số mũ
- Xác định giá trị của đơn thức khi biết giá trị của x
- Phép toán cộng (+) để cộng hai đơn thức cùng bậc

```python
class Monomial:
    def __init__(self, para_coef = 1, para_exp = 2):
        self.__coef = para_coef
        self.__exp = para_exp
    
    def value(self, x):
        return self.__coef*(x**self.__exp)
    
    @property
    def exponent(self):
        return self.__exp
    
    @property
    def coefficient(self):
        return self.__coef
    
    def __add__(self, other):
        if self.__exp == other.exponent:
            coef = self.__coef + other.coefficient
            exp = self.__exp
            return Monomial(coef, exp)
        else: 
            raise ValueError("Cannot add monomials with different exponenents")
    
    def __str__(self):
        return f'P(x) = {self.__coef}x^{self.__exp}'
```

```python
p1 = Monomial(3,2)
p2 = Monomial(7,5)
p3 = Monomial(5,2)

print('p1:', p1)
print('p1(4) =', p1.value(4))
print(p1+p3)
```

Định nghĩa class Money để mô tả thông tin về tiền lưu hành ở châu Âu như sau:

class Money
{

private:

    int iEuro;
    int iCent;

};

Cho biết 1 euro = 100 cents. Trong lớp Money, hãy định nghĩa các phương thức:

a. Toán tử nhập

b. Toán tử cộng hai đối tượng.

```python
class Money:
    def __init__(self, euro=0, cent=0):
        self.__iEuro = euro
        self.__iCent = cent
    
    def input(self):
        self.__iEuro = int(input('Euro:'))
        self.__iCent = int(input('Cent:'))
        
    @property
    def euro(self):
        return self.__iEuro
    
    @property
    def cent(self):
        return self.__iCent
    
    def __add__(self, other):
        sub_euro = 0
        cent = self.__iCent + other.cent
        if cent >= 100:
            sub_euro = cent//100
            cent = cent%100
        euro = self.__iEuro + other.euro + sub_euro
        return Money(euro, cent)
    
    def __str__(self):
        return f'Money: {self.__iEuro} euro {self.__iCent} cent'
    
m1 = Money()
print(m1)

m1.input()
print(m1)

m2 = Money(2,94)
print(m1+m2)
```

```python
class IntArr:
    def __init__(self, cnt = 0, value = None):
        if cnt < 0:
            raise ValueError("The array must have at least 0 elements.")
        if (cnt > 0) and (value is None): 
            value = 0
        self.__count = cnt
        self.__values = [value]*cnt if isinstance(value, int) else list(value or [])
    
    @property
    def count(self):
        return self.__count
    
    @property
    def values(self):
        return self.__values
    
    def concat(self, other):
        count = self.__count + other.count
        value = self.__values + other.values
        return IntArr(count, value)
    
    def push(self, x):
        self.__values = self.__values + [x]
        self.__count += 1
        
    def input(self):
        self.__count = int(input('Enter the number of elements: '))
        self.__values = [int(input(f'Enter element {i+1}: ')) for i in range(self.__count)]
        
    def __str__(self):
        return ' '.join(map(str, self.__values))
    
    def __repr__(self):
        return f'Count: {self.__count}\nArray: {self.__values}'
    

l1 = IntArr()
print(l1)

l2 = IntArr(3,2)
print(l2)

l3 = IntArr(2)
print(l3)

l4 = l2.concat(l3)
print(l4)

l2.push(3)
print(l2)

l2.input()
print(l2)
```

```python
class Fraction:
    def __init__(self, num = 1, deno = 1):
        if deno == 0:
            raise ValueError('Fraction can not have denominator 0')
        self.__numerator = num
        self.__denominator = deno
    
    def input(self):
        self.__numerator = int(input('Enter the numerator: '))
        self.__denominator = int(input('Enter the denominator: '))
        while self.__denominator==0:
            print ('Fraction can not have denominator 0')
            self.__denominator = int(input('Enter the denominator again: '))
    
    @property
    def numerator(self):
        return self.__numerator
    
    @property
    def denominator(self):
        return self.__denominator
    
    def __add__(self, other):
        num = self.__numerator*other.denominator + other.numerator*self.__denominator
        deno = self.__denominator*other.denominator
        return Fraction(num, deno)
    
    def __eq__(self, other):
        GCD1 = Fraction.gcd(self.__numerator, self.__denominator)
        GCD2 = Fraction.gcd(other.numerator, other.denominator)
        return (self.__numerator // GCD1 == other.numerator // GCD2 and
                self.__denominator // GCD1 == other.denominator // GCD2)
        
        
    @staticmethod
    def gcd(a,b):
        a = abs(a)
        b = abs(b)
        while (a!=b):
            if a>b:
                a -= b  
            else: 
                b -= a
        return a
        
    def __str__(self):
        if self.__denominator<0:
            self.__numerator *= -1
            self.__denominator *= -1
        
        if self.__numerator%self.__denominator == 0:
            temp = self.__numerator//self.__denominator
            return f'{temp}'
        
        GCD = Fraction.gcd(self.__numerator, self.__denominator)
        self.__numerator //= GCD
        self.__denominator //= GCD
        return f'{self.__numerator}/{self.__denominator}'
    
    
a = Fraction(5,3)
print(a)

b = Fraction()
b.input()

c = Fraction()
c.input()

res = a+b+5+c
print(res)

if(a==b):
    print('Fraction a is equal to fraction b')
```
