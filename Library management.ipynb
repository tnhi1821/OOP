The school library is the cultural and scientific center of the school aimed at expanding knowledge, contributing to improving the effectiveness of teaching and learning, and scientific research of teachers and students.
A school library needs to manage a list of books, including the following two types:

● Textbooks: book code, book name, publisher, quantity, unit price, and condition (0: old, 1: new).
- If the book condition is new: Total amount = quantity * unit price.
- If the book condition is old: Total amount = quantity * unit price * 50%.

● Reference books: book code, book name, publisher, quantity, unit price, and tax amount.
- Total amount = quantity * unit price + tax

Requirements:
1. Define classes to manage according to the school library's requirements.
2. Enter and output the list of books in the school library.
3. Calculate the total amount the school library has to pay for each type of book and display detailed information on the books with the lowest total amount.
4. Find information on textbooks from the publisher whose name is entered by the user.

Thư viện trường học là trung tâm văn hóa, khoa học của nhà trường nhằm mở rộng kiến thức,
góp phần nâng cao hiệu quả dạy - học và nghiên cứu khoa học của giáo viên và học sinh.
Một thư viện trường học cần quản lý danh sách các loại sách gồm 2 loại như sau:

● Sách giáo khoa: mã sách, tên sách, nhà xuất bản, số lượng, đơn giá và tình trạng (0: cũ, 1:
mới).
- Nếu tình trạng sách là mới thì: Thành tiền = số lượng * đơn giá.
- Nếu tình trạng sách là cũ thì: Thành tiền = số lượng * đơn giá * 50%

● Sách tham khảo: mã sách, tên sách, nhà xuất bản, số lượng, đơn giá và tiền thuế.
- Thành tiền = số lượng * đơn giá + thuế

Yêu cầu:
1. Định nghĩa các lớp để quản lý theo yêu cầu của thư viện trường học.
2. Nhập và xuất danh sách các sách của thư viện trường học.
3. Tính thành tiền mà thư viện trường học phải trả cho từng loại sách và hiển thị thông tin chi tiết những sách nào có thành tiền phải trả ít nhất.
4. Tìm thông tin các sách giáo khoa của nhà xuất bản có tên do người dùng nhập vào.


```python
## Class
class Book:
    def __init__(self, book_id, title, publisher, quantity, price):
        self._book_id = book_id
        self._title = title
        self._publisher = publisher
        self._quantity = quantity
        self._unit_price = price
        
    @property
    def publisher(self):
        return self._publisher
    
    def calculate(self):
        pass
    
    def __str__(self):
        return f"Book ID: {self._book_id}\nTitle: {self._title}\nPublisher: {self._publisher}\nQuantity: {self._quantity}\nUnit Price: {self._unit_price}"
    
class TextBook(Book):
    def __init__(self, book_id, title, publisher, quantity, price, condition):
        super().__init__(book_id, title, publisher, quantity, price)
        self.__condition = condition
    
    def calculate(self):
        if self.__condition == 1:
            return self._quantity * self._unit_price
        else:
            return self._quantity * self._unit_price * 0.5
    
    def __str__(self):
        return f'{super().__str__()}\nCondition: {self.__condition}'
    
class ReferenceBook(Book):
    def __init__(self, book_id, title, publisher, quantity, price, fax):
        super().__init__(book_id, title, publisher, quantity, price)
        self.__fax = fax
    
    def calculate(self):
        return self._quantity * self._unit_price + self.__fax
    
    def __str__(self):
        return f'{super().__str__()}\nFax: {self.__fax}'
    
## Program function
def InputBookList():
    book_list = []

    num_textbooks = int(input("Enter the number of textbooks: "))
    for _ in range(num_textbooks):
        book_id = input("Enter book ID: ")
        title = input("Enter title: ")
        publisher = input("Enter publisher: ")
        quantity = int(input("Enter quantity: "))
        unit_price = float(input("Enter unit price: "))
        condition = int(input("Enter condition (0: old, 1: new): "))
        book = TextBook(book_id, title, publisher, quantity, unit_price, condition)
        book_list.append(book)

    num_reference_books = int(input("Enter the number of reference books: "))
    for _ in range(num_reference_books):
        book_id = input("Enter book ID: ")
        title = input("Enter title: ")
        publisher = input("Enter publisher: ")
        quantity = int(input("Enter quantity: "))
        unit_price = float(input("Enter unit price: "))
        tax = float(input("Enter tax: "))
        book = ReferenceBook(book_id, title, publisher, quantity, unit_price, tax)
        book_list.append(book)
    
    return book_list

def OutputBookList(book_list):
    for book in book_list:
        print(book)
        print('------------------------------')
        
def FindMinUnitPrice(book_list):
    text = 0
    ref = 0
    book_price = []
    
    for book in book_list:
        total_unit_price = book.calculate()
        book_price.append(total_unit_price)
        if isinstance(book, TextBook):
            text += total_unit_price
        else:
            ref += total_unit_price
        
    print(f'Total amount for each type:\nText book: {text}\nReference book: 4{ref}')
    min_price = min(book_price)
    print(f'Min total unit price: {min_price}')
    
    print("Books with the minimum total unit price:")
    for book in book_list:
        if book.calculate() == min_price:
            print(book)
    
def FindTextbook(book_list):
    flag = False
    publisher_name = input('Enter publisher name: ')
    for book in book_list:
        if book.publisher == publisher_name:
            flag = True
            print(book)
    if flag == False:
        print(f'Book list has no book with publisher name {publisher_name}')
        
    
## Main program
while True:
    print('========================================')
    print('CONTROL TABLE')
    print('1. Input book list')
    print('2. Output book list')
    print('3. Find min total unit price')
    print('4. Find textbooks by publisher')
    print('5. Exit')
    print('========================================')
    key = int(input('Enter your option: '))
    
    if key==1:
        book_list = InputBookList()
    elif key==2:
        OutputBookList(book_list)
    elif key==3:
        FindMinUnitPrice(book_list)
    elif key==4:
        FindTextbook(book_list)
    else:
        print('Thank you for using this program!')
        break
```
